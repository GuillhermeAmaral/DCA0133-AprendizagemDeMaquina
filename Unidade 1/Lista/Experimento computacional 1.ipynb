## Gerando uma distribuição Exponencial
Só para refrescar a memória, a distribuição exponencial é a seguinte ([fonte](https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_exponencial)):
$$f(x;\lambda) = 
\begin{cases}
\lambda e^{-\lambda x} &\text{, }  x\geq 0\\
0 &\text{, }  x<0 
\end{cases}
= \begin{cases}
\dfrac{1}{\lambda}e^{\dfrac{-x}{\lambda}} &, x \geq 0, \lambda > 0\\
0 & , x < 0
\end{cases}$$

```python
lamb = 0.5 # Lambda, o parâmetro da distribuição exponencial
va_exp = -1/lamb*np.log(va) # Inversa da função distribuição cumulativa F_x(X), X é uma v.a.
plt.hist(va_exp, 50, density=True) # density=True faz com que o gráfico seja normalizado

x = np.linspace(0, 20, len(va_exp))
y1 = lamb*np.exp(-x*lamb) # função distribuição de probabilidade exponencial, ou f_x(X)
plt.plot(x,y1,'k')
plt.title('Comparação entre a curva obtida e a exata. Lambda='+str(lamb))
plt.show()
```

## Distribuição Qui-Quadrado
A distribuição Qui-Quadrado tem f.d.p. dada por:
$\displaystyle {\frac {1}{2^{k/2}\Gamma (k/2)}}\;x^{k/2-1}e^{-x/2}$

A distribuição pode ser simulada a partir da distribuição normal. Por definição, se $Z_{1}, Z_{2}, ...,Z_{k}$ forem distribuições normais padronizadas independentes, então a soma de seus quadrados é uma Qui-Quadrado com k graus de liberdade.
$$\chi_{k}^{2} = Z_{1}^{2}+Z_{2}^{2}+ ... +Z_{k}^{2}$$

```python
from math import gamma

N = 69999  # Qunatidade de amostras em cada v.a.
grau = 8  # Grau de liberdade
qtd_bins = 800  # Quantidade de barras no histograma

def dist_normal(qtd_va, qtd_amost, media, desvio):
    '''
    Gera uma distribuição  normal customizada

    :param qtd_va: quantidade de variáveis aleatórias
    :param qtd_amost: quantidade de amostras e cada va
    :param media:  media da dist. normal
    :param desvio: desvio dadist. normal
    :return: Retorna um vetor cujos elementos seguem a distribuição
    normal com media e desvio padrão dados
    '''

    va_gauss = np.zeros(qtd_amost)
    for i in range(qtd_va):  # Gerando 1000 variáveis aleatórias
        va = np.random.random(qtd_amost)  # Gera valores aleatórios no intervalo [0, 1)
        va_gauss = va_gauss + va  # Faz o somatório as va's geradas

    media_gauss = va_gauss.mean()  # Média da va_gaussiana
    desvio_gauss = va_gauss.std()  # Desvio padrão da va_gaussiana

    Z = (va_gauss - media_gauss) / desvio_gauss  # Z já tem distrbuição normal padronizada
    Y = Z * desvio + media  # Para dimensionar a dist.
    return Y

def dist_qui(x, k):
    return x**(k/2.0-1)*e**(-x/2.0)/(gamma(k/2)*2**(k/2.0))


qui = np.zeros(N)
for i in range(grau):
    qui = qui + dist_normal(100, N, 0, 1) ** 2

x = np.linspace(0, 30, 200)
y = dist_qui(x, 8)
plt.plot(x, y, 'k')
plt.hist(qui, qtd_bins, density=True)
plt.title('Qui-Quadrado com grau de liberdade '+str(grau))
plt.xlabel('Histograma com '+str(qtd_bins)+' divisões')
plt.show()
```

## Gerando uma distribuição de Poisson (pendente)
A distribuição de Poisson tem a seguinte fórmula:

$$p(x) = P(X=x) = e^{-\lambda}\dfrac{\lambda^{x}}{x!} \hspace{30px} x \in Z \text{, e }\lambda > 0$$

Um algoritmo que pode ser usado chama-se "Método da Aceitação/Rejeição":
<ol>
    <li>Fazer $n=0$ e $P=1$;</li>
    <li>Gerar um número aleatório $R_{n}$ e substituir $P$ por $P\cdot R_{n}$;</li>
    <li>Se $P < e^{-\lambda}$, aceitar X = n e voltar ao passo 1, caso contrário, rejeitar n atual, fazer $n=n+1$ e retornar aos procedimentos no passo 2.</li>
</ol>
